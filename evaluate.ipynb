```python
import pandas as pd
import matplotlib.pyplot as plt
import os
from pathlib import Path

project_path = Path(os.path.abspath('__file__')).parent
log_path = project_path / 'population' / 'log.csv'
print(f'log_path {log_path}')

df = pd.read_csv(log_path)

fig = plt.figure(figsize=(35,10))
ax = plt.subplot()
ax.set_title('Steps', fontsize=16)
plt.errorbar(df.index, df.steps_mean, df.steps_std, fmt='ok', ecolor='green', lw=1)
plt.errorbar(df.index, df.steps_mean, [df.steps_mean-df.steps_min, df.steps_max-df.steps_mean], fmt='.k', ecolor='gray', lw=.5)
plt.xlim(-1, max(df.index)+1)
plt.show()

fig = plt.figure(figsize=(35,10))
ax = plt.subplot()
ax.set_title('Apples', fontsize=16)
plt.errorbar(df.index, df.apples_mean, df.apples_std, fmt='ok', ecolor='green', lw=1)
plt.errorbar(df.index, df.apples_mean, [df.apples_mean-df.apples_min, df.apples_max-df.apples_mean], fmt='.k', ecolor='gray', lw=.5)
plt.xlim(-1, max(df.index)+1)
plt.show()

fig = plt.figure(figsize=(35,10))
ax = plt.subplot()
ax.set_title('Fitness', fontsize=16)
plt.errorbar(df.index, df.fitness_mean, df.fitness_std, fmt='ok', ecolor='green', lw=1)
plt.errorbar(df.index, df.fitness_mean, [df.fitness_mean-df.fitness_min, df.fitness_max-df.fitness_mean], fmt='.k', ecolor='gray', lw=.5)
plt.xlim(-1, max(df.index)+1)
plt.show()
```
